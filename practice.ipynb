```python
##side quest
def sample(l1,l2):
    a = 0
    b = 0
    for i in l1:
        a+=1
    for i in l2:
        b+=1
    for c in range(0,a):
        for d in range(0,b):
            if (l1[c]==l2[d]):
                return 1
    return 0

l1=input("Enter upto 6 numbers with comma given: ").split(",",6)
l2=input("Enter upto 6 numbers with comma given: ").split(",",6)
if (sample(l1,l2)):
    print("overlaping")
else: print("not overlaping")
```

```python
#calculating time using decorators
import time
import math
def cal_time(func):
    def inner1(*args,**kwargs):
        start = time.time()
        func(*args,**kwargs)
        finish = time.time()
        print("Total time taken in: ",func.__name__,finish - start)
    return inner1

@cal_time
def factorial(number):
    # time.sleep(2)
    print(math.factorial(number))

factorial(10)
        
```

```python
#by returning arguments
def main(head):
    def branch(*args,**kwargs):
        print("before operation")
        func = head(*args,**kwargs)
        print("after operation")
        return func
    return branch

@main
def sum_of_two(a,b):
    print("main operation")
    return a+b

a,b = int(input()),int(input())
print("Sum = ",sum_of_two(a,b))
```

```python
#grand inheritance
class Parent:
    def __init__(self,time):
        self.time = time
    
    def dis_time(self):
        return self.time
    
class Child(Parent):
    def __init__(self,time,money):
        Parent.__init__(self,time)
        self.money = money
    
    def dis_money(self):
        return self.money
    
class GrandChild(Child):
    def __init__(self,time,money,skills):
        Child.__init__(self,time,money)
        self.skill = skills
    
    def dis_skill(self):
        return self.skill
    
gc = GrandChild("Full",10000000,"Top-Notch")
print(gc.dis_time(),gc.dis_money(),gc.dis_skill())

```
